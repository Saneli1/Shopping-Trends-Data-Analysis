```python
## Maximum and Minimum Values: Determine the highest and lowest purchase amounts 
##or ages in the dataset. 
file_path = r'C:\Users\user\Downloads\shopping_trends.csv'
with open(file_path, 'r') as file:
    data = file.readlines()
max_purchase = float('-inf')
min_purchase = float('inf')
max_age = float('-inf')
min_age = float('inf')
for row in data[1:]:
    columns = row.strip().split(",")
    purchase_amount = float(columns[5].strip())
    age = float(columns[1].strip())
    if purchase_amount > max_purchase:
       max_purchase = purchase_amount
    if purchase_amount < min_purchase:
        min_purchase = purchase_amount
    if age > max_age:
        max_age = age
    if age < min_age:
        min_age = age
print("Max Purchase Amount:", max_purchase)
print("Min Purchase Amount:", min_purchase)
print("Max Age:", max_age)
print("Min Age:", min_age)
```

```python
max_review_rating = float('-inf')
min_review_rating = float('inf')
for row in data[1:]:
    columns = row.strip().split(",")
    review_rating = float(columns[10].strip())
    if review_rating > max_review_rating:
        max_review_rating = review_rating
    if review_rating < min_review_rating:
        min_review_rating = review_rating
print("Max Review Rating:", max_review_rating)
print("Min Review Rating:", min_review_rating)
```

```python
##Average age 
tot_age = 0
count_age = 0
for row in data[1:]:
    columns = row.strip().split(",")
    try:
        age = float(columns[1].strip())
        tot_age += age
        count_age += 1
    except ValueError:
        print(f"Skipping row: {columns[1]}")
if count_age > 0:
    avg_age = tot_age / count_age
    print(f"Avg Age is {avg_age}")
else:
    print("No Avg age")
```

```python
##Avg purchase amount
tot_purchase = 0
count_purchase = 0
for row in data[1:]:
    columns = row.strip().split(",")
    try:
        purchase_amount = float(columns[5].strip())
        tot_purchase += purchase_amount
        count_purchase += 1
    except ValueError:
        print(f"Skipping row: {columns[5]}")
if count_purchase > 0:
    avg_purchase_amount = tot_purchase / count_purchase
    print(f"Avg Purchase Amount is: {avg_purchase_amount}")
```

```python
tot_location_counts = {}
count_location_counts = {}
for row in data [1:]:
    columns = row.strip().split(",")
    location = columns[6].strip()
    purchase_amount = float(columns[5].strip())
    if location in tot_location_counts:
        tot_location_counts[location] += purchase_amount
        count_location_counts[location] += 1
    else:
        tot_location_counts[location] = purchase_amount
        count_location_counts[location] = 1
avg_location_counts = {}
for location in tot_location_counts:
    avg_location_counts[location] = tot_location_counts[location] / count_location_counts[location]

print(avg_location_counts)
        
```

```python
import pandas as pd
df = pd.read_csv(r'C:\Users\user\Downloads\shopping_trends.csv')
print(df.head())
```

```python
## create a frequency count
season_counts = df["Season"].value_counts()
print(season_counts)
```

```python
##summarize tot purchase amount by category
category_summary = df.groupby("Category")["Purchase Amount (USD)"].sum()
print(category_summary)
```

```python
##visualize data
import matplotlib.pyplot as plt
season_counts.plot(kind = "bar", color = "blue")
plt.title("Number of Purchases per Season")
plt.xlabel("Season")
plt.ylabel("Count")
plt.show()
```

```python
##plot total purchase amount by category
category_summary.plot(kind = "bar", color = "lightblue")
plt.title("Number of purchases by the category")
plt.xlabel("Category")
plt.ylabel("Total Purchase Amount (USD)")
plt.show()
```
